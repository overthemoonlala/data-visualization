# 타이타닉 데이터 분석 및 시각화

PassengerID : 탑승객 고유 아이디

Survival : 탑승객 생존 유무 (0: 사망, 1: 생존)

Pclass : 등실의 등급

Name : 이름

Sex : 성별

Age : 나이

Sibsp : 함께 탐승한 형제자매, 아내, 남편의 수

Parch : 함께 탐승한 부모, 자식의 수

Ticket :티켓 번호

Fare : 티켓의 요금

Cabin : 객실번호

Embarked : 배에 탑승한 항구 이름 ( C = Cherbourn, Q = Queenstown, S = Southampton)


```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.dropna().describe() #누락값 없애주기
```

```python
titanic.var()
```

```python
titanic.mad() #평균절대편차
```

```python
titanic.groupby('class').std()
```

```python
titanic.groupby('class')['fare'].median()
```

```python
titanic.query("alive == 'yes'").groupby('class').count()   #''빼먹지 말기~~!!
```

```python
titanic.groupby('class')['age'].describe() #min  25  50  75 max,class별로 묶고 나이!!
```

```python
titanic.query("alive == 'yes'").groupby('class').describe()
```

```python
titanic.groupby('sex')['age'].aggregate([min, np.median, max]) # aggregate(): 요약 통계 값 반환
```

```python
titanic.query('age > 30 ').groupby('class').median() #Fist에서 가장 많은 나이대 탑승
```

```python
titanic.groupby(['class', 'sex'])['age'].mean().unstack() #class와 성별에 따라!
```

unstack(): 데이터프레임 재구조

image.png

```python
sns.catplot(x='who',
            y = 'age',
            hue = 'class',
            kind = 'bar',
            data= titanic) # catplot은 data인자에 데이터프레임 구조를 갖는 데이터를 넣어주고, x, y인자에 칼럼명을 설정
```

```python
titanic.groupby(['class', 'sex'])['fare'].mean().unstack() #운임별
```

```python
titanic.groupby(['class', 'who'])['fare'].mean().unstack() # 요금
# => 남자 운용비용저렴~
# => 아이> 여자> 남자
```

```python
sns.catplot(x = 'sex',
            y='fare',
            hue='class',
            kind='bar',
            data= titanic)
```

```python
sns.catplot(x = 'sex',
            y='fare',
            hue='class',
            kind='bar',
            data= titanic)
```

```python
titanic.groupby(['class', 'sex'])['survived'].mean().unstack()
```

```python
titanic.pivot_table('survived', index = 'class', columns = 'who')
```

```python
sns.catplot(x='class', y='survived', hue='sex', kind='bar', data= titanic)
```

```python
age = pd.cut(titanic['age'], [0,18, 40, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
fare = pd.qcut(titanic['fare'], 3)
titanic.pivot_table('survived', ['who', age],[fare, 'class'])
```

```python
titanic.pivot_table('survived', index = 'who', columns = 'class', margins = True)
```

```python
sns.catplot(x='class', y='survived',hue='who', kind='bar', data=titanic);
```

```python
sns.catplot(x='class',
            y='survived',
            col='who',
            kind = 'bar',
            data = titanic);
```

```python
titanic.pivot_table('survived', index = 'deck', columns='class', margins= True)
```

```python
sns.countplot(y='deck', data = titanic);
```

```python
sns.countplot(y='deck', hue='class', data = titanic);
```

```python
sns.catplot(x='survived', y='deck', hue='class', kind='bar', data= titanic);
```

```python
titanic.pivot_table('survived', index ='embark_town', columns='class', margins = True)
```

```python
sns.countplot(y='embark_town', data= titanic);
```

```python
sns.catplot(x='survived', y= 'embark_town', hue='class', kind= 'bar', data = titanic);
# 생존의 여부와 상관이 없움
```

```python
sns.catplot(x='sibsp', y='survived', kind='bar', data = titanic);
```
